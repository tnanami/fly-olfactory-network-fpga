```python
import serial
import time
import threading
import matplotlib.pyplot as plt
import os
import platform
import psutil
import csv
import numpy as np
import pandas as pd
import random
import seaborn as sns
import statistics
import math
from matplotlib.ticker import ScalarFormatter
```

```python
fig = plt.figure(constrained_layout=True, figsize=(10,4))
subfigs = fig.subfigures(1, 2,wspace=0.05, width_ratios=[1, 1])
ax0 = subfigs[0].add_subplot(1,1,1)
b0=np.zeros(5)
b1=np.zeros(5)
b2=[]
b0=[9609,7669,47.5,10,1]
b1[0]=20800
b1[1]=41600
b1[2]=50
b1[3]=90
b1[4]=5
b2.append("LUT")
b2.append("FF")
b2.append("BRAM")
b2.append("DSP")
b2.append("MMCM")
w0=0.6
cmap = plt.get_cmap("tab20")
for i in range(5):
    ax0.bar(b2[i],(b0[i]/b1[i]*100),color=cmap(i),width=w0)
    ax0.text(b2[i],(b0[i]/b1[i]*100) , b0[i], ha='center', va='bottom')
ax0.set_ylim(0,100)
ax0.set_ylabel("Resource utilization [%]")
ax0.text(-0.15, 0.9, "a", ha='left',fontsize=24, fontweight='bold', transform=ax0.transAxes)

ax1 = subfigs[1].add_subplot(1,1,1)
c0=["Clock", "Signal", "Logic", "BRAM", "DSP", "MMCM", "I/O", "Static"]
c1=[0.03, 0.048, 0.043, 0.059, 0.01, 0.104, 0.001, 0.075]
c2=[cmap(8),cmap(9),cmap(10),cmap(2),cmap(3),cmap(4),cmap(6),cmap(7)]
print(sum(c1))
ax1.pie(list(map(lambda x: x/np.sum(c1) ,c1)), labels=c0, counterclock=False, startangle=90, autopct="%1.1f%%",colors=c2)

ax1.set_ylabel("Power estimation")
ax1.text(-0.1, 0.9, "b", ha='left',fontsize=24, fontweight='bold', transform=ax1.transAxes)
centre_circle = plt.Circle((0,0),0.3,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
ax1.text(0.5, 0.45, "total\n0.37W", ha='center',fontsize=12, transform=ax1.transAxes)
plt.savefig("../outputs/figs/power.svg", format="svg",dpi=1200)
```
